### variable explanations
**CO2168**: Carbon monoxide concentration in ppb-units at 2168 measurement level (216.8 m) above Mast base  
**Glob**: Global shortwave solar radiation in wavelength range 0.3-4.8 µm  
**H2O168**: Water vapour concentration at 16.8 m height  
**NET**: Net radiation  
**NO168**: Nitrogen monoxide concentration in dry air at 16.8 m height  
**NOx168**: Nitrogen oxide concentration in dry air at 16.8 m height  
**O3168**: Ozone concentration in dry air at 16.8 m height  
**Pamb0**： Atmospheric pressure at ground level (180 m a.s.l.)  
**PAR**: Photosynthetically active radiation in wavelength range 400-700 nm  
**PTG**: Potential temperature gradient  
**RGlob**: Reflected shortwave radiation  
**RHIRGA168**: Relative humidity calculated from H₂O concentration and air temperature at 16.8 m height  
**RPAR**: Reflected photosynthetically active radiation in wavelength range 400-700 nm at 67 m height in the mast  
**SO2168**: Sulphur dioxide concentration in dry air at 16.8 m height  
**SWS**: Surface wetness sensor signal  
**T168**: Air temperature at 16.8 m height measured with Pt100 inside ventilated custom-made radiation shield  
**UV_A**: Ultraviolet A radiation in wavelength range 320-400 nm at 18 m height  
**UV_B**: Ultraviolet B radiation in wavelength range 280-320 nm at 18 m height  
**CS**: Condensation sink in the units of 1/s
