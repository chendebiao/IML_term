```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

npf = pd.read_csv("data/npf_train.csv")


npf = npf.set_index("date")

## Tell that "class4" is categorical variable. (R does this automatically.)
npf["class4"] = npf["class4"].astype("category")

npf = npf.drop(["id","partlybad"],axis=1)
corr = npf.corr(method='pearson')
```

```python
CO2_corr =corr.loc[corr.index.str.startswith(('CO2'))&corr.index.str.endswith(('mean')),\
    corr.columns.str.startswith(('CO2'))&corr.columns.str.endswith(('mean'))]

CO2_corr.min()
```

```python
radiation_corr = corr.loc[corr.index.str.startswith(('Glob','NET','PAR','RGlob','RPAR','UV')),\
    corr.columns.str.startswith(('Glob','NET','PAR','RGlob','RPAR','UV'))]

print(radiation_corr.mean(axis=1))
print(radiation_corr.min(axis=1))
radiation_corr
```
